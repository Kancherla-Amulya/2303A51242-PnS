```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

```

```python
# Given data
advertising_budget = [1000, 2000, 3000, 4000, 1500, 2500, 3500, 1200, 4500, 5000]
sales_revenue = [5000, 7000, 8000, 9500, 6000, 7500, 8700, 5500, 9900, 11000]
social_media_engagement = [200, 250, 300, 400, 220, 280, 350, 210, 450, 500]
```

```python
# Scatter plots
plt.figure(figsize=(15, 5))

```

```python
# Advertising Budget vs. Sales Revenue
plt.subplot(1, 3, 1)
plt.scatter(advertising_budget, sales_revenue, color='blue')
plt.title("Advertising Budget vs. Sales Revenue")
plt.xlabel("Advertising Budget")
plt.ylabel("Sales Revenue")

```

```python
# Social Media Engagement vs. Sales Revenue
plt.subplot(1, 3, 2)
plt.scatter(social_media_engagement, sales_revenue, color='green')
plt.title("Social Media Engagement vs. Sales Revenue")
plt.xlabel("Social Media Engagement")
plt.ylabel("Sales Revenue")

```

```python
# Advertising Budget vs. Social Media Engagement
plt.subplot(1, 3, 3)
plt.scatter(advertising_budget, social_media_engagement, color='red')
plt.title("Advertising Budget vs. Social Media Engagement")
plt.xlabel("Advertising Budget")
plt.ylabel("Social Media Engagement")

plt.tight_layout()
plt.show()
```

```python
# Pearson Correlation Coefficients
# Advertising Budget vs. Sales Revenue
corr1, _ = pearsonr(advertising_budget, sales_revenue)
print(f"Correlation between Advertising Budget and Sales Revenue: {corr1:.2f}")

# Social Media Engagement vs. Sales Revenue
corr2, _ = pearsonr(social_media_engagement, sales_revenue)
print(f"Correlation between Social Media Engagement and Sales Revenue: {corr2:.2f}")

# Advertising Budget vs. Social Media Engagement
corr3, _ = pearsonr(advertising_budget, social_media_engagement)
print(f"Correlation between Advertising Budget and Social Media Engagement:{corr3:.2f}")
```
