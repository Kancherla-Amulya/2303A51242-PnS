```python
class Queue:
    def __init__(self):
        self.items=[]
    def enqueue(self, item):
        # New items are added to the front of the list
        self.items.insert(0, item)
    def dequeue(self):
        if not self.isEmpty():
            # If items are added to the front, remove from the back to maintain FIFO
            return self.items.pop()
        return None
    def isEmpty(self):
        return self.items == []
    def size(self):
        return len(self.items)
```

```python
class Q:
  def __init__(self):
    self.q = [] # Initialize q as an instance variable

  def enqueue(self, data):
    self.q.insert(0, data) # Use the instance variable self.q

  def dequeue(self):
    if self.q: # Check if the queue is not empty
      return self.q.pop() # Use the instance variable self.q and pop from the end for FIFO
    return None # Return None if queue is empty

  def get_queue(self):
    return self.q # Added for debugging/inspection

queue=Q()
queue.enqueue(10)
queue.enqueue(40)
queue.enqueue(50)
queue.enqueue(60)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedListQueue:
    def __init__(self):
        self.front = None
        self.rear = None
        self._size = 0

    def is_empty(self):
        return self.front is None

    def enqueue(self, item):
        new_node = Node(item)
        if self.is_empty():
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
        self._size += 1

    def dequeue(self):
        if self.is_empty():
            return None  # Queue is empty

        dequeued_item = self.front.data
        self.front = self.front.next

        if self.front is None: # If the queue becomes empty after dequeue
            self.rear = None

        self._size -= 1
        return dequeued_item

    def peek(self):
        if self.is_empty():
            return None
        return self.front.data

    def size(self):
        return self._size

    def display(self):
        current = self.front
        elements = []
        while current:
            elements.append(current.data)
            current = current.next
        print("Queue: ", elements)
```

```python
from collections import deque
def print_reverse_using_deque(numbers):
    my_queue = deque()
    print(f"Original numbers: {numbers}")
    for num in numbers:
        my_queue.append(num)
    print(f"Queue after enqueuing: {list(my_queue)}")
    my_stack = deque() # deque can also act as a stack
    while my_queue:
        item = my_queue.popleft() # This is the dequeue operation
        my_stack.append(item) # Push onto stack
    print(f"Queue after dequeuing to stack: {list(my_queue)}")
    print(f"Stack after pushing from queue: {list(my_stack)}")
    reversed_numbers = []
    while my_stack:
        reversed_numbers.append(my_stack.pop()) # Pop from stack
    print(f"Reversed numbers using dequeue and stack: {reversed_numbers}")
given_numbers = [10, 20, 30, 40, 50]
print_reverse_using_deque(given_numbers)
```
