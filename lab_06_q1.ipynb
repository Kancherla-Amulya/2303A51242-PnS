```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def generate_customer_data(N, min_visitors=0, max_visitors=10):
    return np.random.randint(min_visitors, max_visitors + 1, size=N)

def compute_pmf(data):
    unique, counts = np.unique(data, return_counts=True)
    pmf = counts / len(data)
    return unique, pmf

def compute_cdf(data):
    unique, counts = np.unique(data, return_counts=True)
    cdf = np.cumsum(counts) / len(data)
    return unique, cdf

# Parameters
N = 365  # Simulating data for one year
data = generate_customer_data(N)

# Compute PMF and CDF
x_pmf, y_pmf = compute_pmf(data)
x_cdf, y_cdf = compute_cdf(data)

# Visualization
sns.set(style="darkgrid")

plt.figure(figsize=(12, 5))

# Plot PMF
plt.subplot(1, 2, 1)
plt.bar(x_pmf, y_pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Customers")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x_cdf, y_cdf, marker='o', linestyle='-', color='red')
plt.xlabel("Number of Customers")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")

plt.tight_layout()
plt.show()
```
